```python
import numpy as np
np.random.seed(0)

from scipy.stats import beta

import matplotlib.pyplot as plt
import seaborn as sns
import tikzplotlib
```

```python
xs = np.linspace(0, 1, 1000)

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].plot(xs, beta.pdf(xs, a=5, b=2))
ax[1].plot(xs, beta.pdf(xs, a=1, b=1))

ax[0].set_xlabel('\n' + r'$\theta$' + ' is relatively large', fontsize=15)
ax[0].set_ylabel(r'$p(\theta)$', fontsize=15)
ax[1].set_xlabel('\nNo prior knowledge', fontsize=15)

# plt.show()
tikzplotlib.save('betas.tex')
```

```python
samples = np.random.binomial(1, 0.3, size=100)

plt.plot(xs, beta.pdf(xs, a=5, b=2), label='prior')
plt.hist(samples, label='samples', density=True, alpha=0.3)
plt.plot(xs, beta.pdf(
    xs, 
    a=5 + (samples == 1).sum(), 
    b=2 + (samples == 0).sum()
), label='posterior')

plt.xlim((0, 1))
plt.legend()
# plt.show()
tikzplotlib.save('prior_post.tex')
```
