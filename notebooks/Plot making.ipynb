```python
import numpy as np
np.random.seed(0)

from scipy.stats import beta

import matplotlib.pyplot as plt
import seaborn as sns
import tikzplotlib
```

```python
xs = np.linspace(0, 1, 1000)

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].plot(xs, beta.pdf(xs, a=5, b=2))
ax[1].plot(xs, beta.pdf(xs, a=1, b=1))

ax[0].set_xlabel('\n' + r'$\theta$' + ' is relatively large', fontsize=15)
ax[0].set_ylabel(r'$p(\theta)$', fontsize=15)
ax[1].set_xlabel('\nNo prior knowledge', fontsize=15)

# plt.show()
tikzplotlib.save('betas.tex')
```

```python
samples = np.random.binomial(1, 0.3, size=100)

plt.plot(xs, beta.pdf(xs, a=5, b=2), label='prior')
plt.hist(samples, label='samples', density=True, alpha=0.3)
plt.plot(xs, beta.pdf(
    xs, 
    a=5 + (samples == 1).sum(), 
    b=2 + (samples == 0).sum()
), label='posterior')

plt.xlim((0, 1))
plt.legend()
# plt.show()
tikzplotlib.save('prior_post.tex')
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor as GPR
from sklearn.gaussian_process.kernels import Matern
```

```python
xs = np.linspace(1, 10, 1000)
eta = 3
l = 1

model = GPR(eta**2 * Matern(l, nu=2.5), optimizer=None)
```

```python
mean, sd = model.predict(xs.reshape(-1, 1), return_std=True)
```

```python
plt.plot(xs, mean, label='mean')
plt.fill_between(
    xs, mean + 1.96 * sd, mean - 1.96 * sd, 
    alpha=.2, label='95% CI'
)

plt.ylim(-10, 10)
plt.legend()

plt.show()
# tikzplotlib.save('gp_prior.tex')
```

```python
def gramacy_and_lee(x):
    return 5 * np.sin(np.pi * x) / x + (x / 10 - 1) ** 2


np.random.seed(1)
X_sample = np.random.uniform(low=1, high=10, size=10).reshape(-1, 1)
Y_sample = gramacy_and_lee(X_sample)

plt.plot(xs, gramacy_and_lee(xs), c='k', linestyle='--', label='latent function')
plt.scatter(X_sample, Y_sample, c='r', label='obs')
plt.ylim(-10, 10)
plt.legend()

tikzplotlib.save('gp_latent.tex')
```

```python
model.fit(X_sample, Y_sample)
mean, sd = model.predict(xs.reshape(-1, 1), return_std=True)
```

```python
plt.plot(xs, mean, label='mean')
plt.fill_between(
    xs, mean.flatten() + 1.96 * sd, mean.flatten() - 1.96 * sd, 
    alpha=.2, label='95% CI'
)

plt.scatter(X_sample, Y_sample, c='r')

plt.ylim(-10, 10)
plt.legend()

# plt.show()
tikzplotlib.save('gp_post.tex')
```

```python
hyperparams = [
    (1, 3), (0.5, 3),
    (1, 10), (0.5, 10)
]

fig, ax = plt.subplots(2, 2, figsize=(20, 15), sharex=True, sharey=True)

for i in range(4):
    tmp_ax = ax[i // 2][i % 2]
    l, eta = hyperparams[i]

    model = GPR(eta**2 * Matern(l, nu=5.2), optimizer=None)
    model.fit(X_sample, Y_sample)
    mean, sd = model.predict(xs.reshape(-1, 1), return_std=True)
    
    tmp_ax.plot(xs, mean, label='mean')
    tmp_ax.fill_between(
        xs, mean.flatten() + 1.96 * sd, mean.flatten() - 1.96 * sd, 
        alpha=.2, label='95% CI'
    )

    tmp_ax.scatter(X_sample, Y_sample, c='r', label='obs')

    tmp_ax.set_ylim(-10, 10)
    tmp_ax.set_title(f'l = {l}, eta = {eta}')

ax[0][0].legend()

# plt.show()
tikzplotlib.save('gp_hyperparams.tex')
```
